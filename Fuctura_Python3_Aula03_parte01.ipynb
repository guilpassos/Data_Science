```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#O que já sabemos?
empresas = ['Apple', 'Google', 'Amazon',
            'Microsoft', 'Tencent', "McDonald's",
            'Visa', 'Facebook', 'Alibaba', 'Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646,
           611.460, 214.023, 196.526,
           191.032, 186.421, 169.966,
           124.273]
           
```

```python
#já é possivel plotar, só com as variaveis
plt.plot(empresas, bilhoes)
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
#Redimnesionamento
plt.rcParams["figure.figsize"] = (15, 8)
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
plt.show()
```

```python
#conhecendo o frontsize
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.show()
```

```python
#mudando a cor e legendado
plt.plot(empresas, bilhoes, color = 'green')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.show()
```

```python
#aumentar a espessura da linha lw
plt.plot(empresas, bilhoes, color = 'green', lw = 5)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.show()
```

```python
#adicionar um marcador
plt.plot(empresas, bilhoes, color = 'blue', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.show()
```

```python
#adicionando linhas de grade
plt.plot(empresas, bilhoes, color = 'blue', marker = 'o')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
#mudando o tipo do gráfico e colocando e legenda acima (plt.bar)
plt.bar(empresas, bilhoes, color = 'blue')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'high right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
#gráfico mais simples
plt.bar(empresas, bilhoes)
```

```python
cores = ['red',  'yellow', 'green', 'blue', 'purple', 'orange', 'magenta', 'white', 'brown', 'black']
```

```python
cores2 = ['dimgray', 'dimgrey', 'gray', 'grey', 'darkgray', 'darkgrey', 'silver', 'lightgray', 'lightgrey', 'gainsboro']
```

```python
plt.bar(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'high right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
plt.bar(empresas, bilhoes, color = cores2, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'high right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
#estilo do gráfico
plt.style.available
```

```python
plt.style.use('grayscale')
plt.bar(empresas, bilhoes, color = cores2, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'high right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
#barh
plt.style.use('grayscale')
plt.barh(empresas, bilhoes, color = cores2, edgecolor='black')
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões de dólares', fontsize=20)
plt.legend(['Bilhões de dólares'], loc = 'high right', fontsize=15)
plt.grid(True)
plt.show()
```
